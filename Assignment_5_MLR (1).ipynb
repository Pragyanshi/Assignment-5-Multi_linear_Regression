# Preparing a model

```python
import statsmodels.formula.api as smf
model=smf.ols('Profit~RDS+ADMS+MKS',data=df1).fit()
```

```python
# Finding Coefficient parameters
model.params
```

```python
# Finding tvalues and pvalues
print(model.tvalues,'\n',model.pvalues)
```

```python
# Finding rsquared values
(model.rsquared,model.rsquared_adj)
```

```python
# Build SLR and MLR models for insignificant variables 'ADMS' and 'MKS'
# Also find their tvalues and pvalues
```

```python
ml_a=smf.ols('Profit~ADMS',data=df1).fit()
#t and p-values
print(ml_a.tvalues,'\n',ml_a.pvalues)
```

```python
ml_m=smf.ols('Profit~MKS',data=df1).fit()
#t and p-values
print(ml_m.tvalues,'\n',ml_m.pvalues)
```

```python
ml_am=smf.ols('Profit~ADMS+MKS',data=df1).fit()
#t and p-values
print(ml_am.tvalues,'\n',ml_am.pvalues)
```

# Correlation Analysis

```python
df1.corr()
```

```python
#format the plot background and scatter plots for all the variables
sns.set_style(style='darkgrid')
sns.pairplot(df1)
```

# Model Validation
# Two Techniques: 1. Collinearity Check & 2. Residual Analysis

```python
# 1) Collinearity Problem Check
# Calculate VIF = 1/(1-Rsquare) for all independent variables
```

```python
rsq_a=smf.ols('ADMS~RDS+MKS',data=df1).fit().rsquared
vif_a =1/(1-rsq_a)
```

```python
rsq_m=smf.ols('MKS~RDS+ADMS',data=df1).fit().rsquared
vif_m =1/(1-rsq_m)
```

```python
rsq_r=smf.ols('RDS~ADMS+MKS',data=df1).fit().rsquared
vif_r =1/(1-rsq_r)
```

```python
#Storing vif values in a data frame
d1={'Variables':['ADMS','RDS','MKS'],'vif':[vif_a,vif_r,vif_m]}
vif_frame = pd.DataFrame(d1)
vif_frame
```

```python
# None variable has VIF>20, No Collinearity, so consider all varaibles in Regression equation
```

# Subset selection

```python
# Build model with ADMS
import statsmodels.formula.api as smf
model1 = smf.ols('Profit~ADMS+RDS',data=df1).fit()
f'AIC:{model1.aic},rsq_model1:{model.rsquared}'
```

```python
# Build model with MKS
import statsmodels.formula.api as smf
model2 = smf.ols('Profit~MKS+RDS',data=df1).fit()
f'AIC:{model2.aic},rsq_model2:{model.rsquared}'
```

# Model Deletion Diagnostics (checking Outliers or Influencers)
## Two Techniques : 1. Cook's Distance & 2. Leverage value

```python
# 1. Cook's Distance: If Cook's distance > 1, then it's an outlier
# Get influencers using cook's distance
from statsmodels.graphics.regressionplots import influence_plot
model_influence = model.get_influence()
(c,_)=model.get_influence().cooks_distance
c
```

```python
#plot the influencers value using stem plot
fig = plt.subplots(figsize=(20,7))
plt.stem(np.arange(len(df1)),np.round(c,3))
plt.xlabel('Row index')
plt.ylabel('Cooks Distance')
plt.show()
```

```python
# Index and value of influencer where C>0.5
np.argmax(c) , np.max(c)
```

# High influence points

```python
df1.shape
```

```python
# Leverage Cuttoff Value = 3*(k+1)/n ; k = no.of features/columns & n = no. of datapoints
k=df1.shape[1]
n=df1.shape[0]
leverage_cutoff = (3*(k+1))/n
leverage_cutoff
```

```python
# 2. Leverage Value using High Influence Points : Points beyond Leverage_cutoff value are influencers
influence_plot(model)
plt.show()
```

```python
df1.iloc[[49]]
```

# Improving the Model

```python
 # Discard the data points which are influencers and reassign the row number (reset_index(drop=True))
df2=df1.drop(df1.index[[49]],axis=0).reset_index(drop=True)
df2
```

# Build model

```python
final_ml_m=smf.ols('Profit~RDS+ADMS+MKS',data=df2).fit()
```

```python
(final_ml_m.rsquared,final_ml_m.aic)
```

```python
final_ml_a=smf.ols('Profit~ADMS+RDS',data=df2).fit()
```

```python
(final_ml_a.rsquared,final_ml_a.aic)
```

Comparing above R-square and AIC values model 'final_ml_m' has high R-square and low AIC value hence include variable 'MKS' so that multi collinearity problem.

```python
(final_ml_m.rsquared,final_ml_m.aic)
```

# Predicting for new data

```python
# say New data for prediction is
new_data=pd.DataFrame({'RDS':70000,"ADMS":90000,"MKS":140000},index=[0])
new_data
```

```python
# Manual Prediction of Price
final_ml_m.predict(new_data)
```
